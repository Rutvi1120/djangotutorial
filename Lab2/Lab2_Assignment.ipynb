```python
# 1. Given the participants' score sheet for your University Sports Day, you are
#     required to find the runner-up n score. You are given scores. Store
# them in a list and find the score of the runner-up.

n=int(input("Enter n : "))
scores=[]
print("Enter Scores : ")
for i in range(n) :
    temp=int(input())
    scores.append(temp)
top=max(scores)
while top in scores :
    scores.remove(top)
print("Score of the Runner-Up : ",max(scores))

```

```python
# 2. Write a Python script to print a dictionary where the keys are numbers between
#     1 and 15 (both included) and the values are square of keys.

mydict={}
for i in range(1,16) :
    mydict[i]=i*i
print("Dictonary having 1 to 15 as keys and their squares as value :\n")
print(mydict)
```

```python
# 3. Create a list of intergers and print squares of all numbers using list
#     Comprehension.
n=int(input("Enter n : "))
print("Enter numbers : ")
num=[]
for i in range(n) :
    num.append(int(input()))

squares=[val ** 2 for val in num]
print("Numbers : ",num);
print("Squares : ",squares);
```

```python
# 4. Create a dictionary of country and their capital. Take n inputs from
# the user. After the dictionary is created, perform the following operations :
# - Search based on a key
# - Delete an element based on a key

n=int(input("Enter n : "))
world={}
for i in range(n) :
    print("Enter Country {} : ".format(i+1))
    country=input()
    print("Enter Capital {} : ".format(i+1))
    cap=input()
    world[country]=cap
print(world)
key=input("Enter country whose capital you want to know : ")
print("Capital of {} is {}".format(key,world[key]))
key=input("Enter country you want to delete : ")
world.pop(key)
print("After deletion : ",world)
```

```python
# 5.Write a Python program to count the number of characters (character
# frequency) in a string.

s=input("Enter the String : ")
freq={}
for ch in s :
    if ch in freq :
        freq[ch]+=1
    else :
        freq[ch]=1
print(freq)
```

```python
# 6. Write a Python program to convert a given string into a list of words,
# modify one word and convert it back to the string.

s=input("Enter the String : ")
lst=s.split()
print(lst)
old=input("Enter the word you want to change : ")
if old in lst :
    index=lst.index(old)
    newword=input("Enter the word you want to replace with : ")
    lst.remove(old)
    lst.insert(index,newword)
    print(lst)
    s=""
    for w in lst :
        s=s+w+" "
    print(s)
else :
    print("Word Not Found")
```

```python
# 7. Write a Python program to count the occurrences of each word in a
# given sentence.

s=input("Enter the String : ")
lst=s.split()
freq={}
for word in lst :
    if word in freq :
        freq[word]+=1
    else :
        freq[word]=1
print(freq)
```
