```python
#1. Calculate the sum of all numbers from 1 to a given number

n = int(input("Enter the number : "))
sum=0
for i in range (1,n+1):
        sum+=i
print("Sum from 1 to {} is {}".format(n,sum))
```

```python
#2. Write a program to display all prime numbers within a range


n = int(input("Enter the number : "))


for i in range (2,n+1):
        isprime=1
        for j in range(2,int(i/2)+1):
                if i%j==0:
                        isprime=0
                        break;
        if isprime==1:
                print(i)
```

```python
# 3. A shop will give discount of 10% if the cost of purchased quantity is more
# than 1000. Ask user for quantity Suppose, one unit will cost 100. Judge and
# print total cost for user.


n=int(input("Enter the quantity : "))
cost=0
if n>1000:
    cost=n*100
    cost-=int(cost*0.1)
else:
    cost=n*100
print("Total cost = "+str(cost))
```

```python
# 4. A school has following rules for grading system:
# a. Below 25 - F
# b. 25 to 45 - E
# c. 45 to 50 - D
# d. 50 to 60 - C
# e. 60 to 80 - B
# f. Above 80 - A
# Ask user to enter marks and print the corresponding grade.


marks=int(input("Enter your marks : "))
grade=''
if marks>80:
    grade='A'
elif marks>60 and marks<=80:
    grade='B'
elif marks>50 and marks<=60:
    grade='C'
elif marks>45 and marks<=50:
    grade='D'
elif marks>25 and marks<=25:
    grade='E'
else:
    grade='F'
print("Grade : "+grade)
```
